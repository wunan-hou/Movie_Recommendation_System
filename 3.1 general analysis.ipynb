# General Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib import cm
import matplotlib
```

```python
a = pd.read_csv('data',index_col = 'Unnamed: 0')
a.set_index('index',inplace = True)
```

```python
# use the same function defined in data cleaning part
def dataframe_to_dict(df):
    # the function is used to transfer columns type read from csv file
    # to the proprate type 
    # thus we can use in later work
    dict_ = dict()
    for i in df.index:
        dict_[i] = dict()
        if pd.notnull(df['director'][i]):
            dict_[i]['director'] = [i.strip("''") for i in df['director'][i][1:-1].split(', ')]
        else:
            dict_[i]['director'] = 'NaN'
        if pd.notnull(df['genres'][i]):
            dict_[i]['genres'] = [i.strip("''") for i in df['genres'][i][1:-1].split(', ')]
        else:
            dict_[i]['genres'] = 'NaN'
        if pd.notnull(df['keywords'][i]):
            dict_[i]['keywords'] = [i.strip("''") for i in df['keywords'][i][1:-1].split(', ')]
        else:
            dict_[i]['keywords'] = 'NaN'
        if pd.notnull(df['rating_value'][i]):
            dict_[i]['rating_value'] = float(df['rating_value'][i])
        else:
            dict_[i]['rating_value'] = 0
        if pd.notnull(df['release_country'][i]):
            dict_[i]['release_country'] = [i.strip("''") for i in df['release_country'][i][1:-1].split(', ')]
        else:
            dict_[i]['release_country'] = 'NaN'
        if pd.notnull(df['release_date'][i]):
            dict_[i]['release_date'] = df['release_date'][i]
        else:
            dict_[i]['release_date'] = 'NaN'
        if pd.notnull(df['reviews'][i]) and df['reviews'][i] != []:
            dict_[i]['reviews'] = df['reviews'][i].replace('"',"\'").strip("['").strip("]'").split("', '")
            dict_[i]['reviews_number'] = len(dict_[i]['reviews'])
        else:
            dict_[i]['reviews'] = 'NaN'
            dict_[i]['reviews_number'] = 0
        if pd.notnull(df['stars'][i]):
            dict_[i]['stars'] = [i.strip("''") for i in df['stars'][i][1:-1].split(', ')]
        else:
            dict_[i]['stars'] = 'NaN'
        if pd.notnull(df['storyline'][i]):
            dict_[i]['storyline'] = df['storyline'][i]
        else:
            dict_[i]['storyline'] = 'NaN'
        if pd.notnull(df['time'][i]):
            dict_[i]['time'] = df['time'][i]
        else:
            dict_[i]['time'] = 'NaN'
        if pd.notnull(df['writers'][i]):
            dict_[i]['writers'] = [i.strip("''") for i in df['writers'][i][1:-1].split(', ')]
        else:
            dict_[i]['writers'] = 'NaN'
    return dict_
```

```python
new_dict = dataframe_to_dict(a)
```

```python
def not_movie_name(dict_,director=None,genres=None,keywords=None,rating_value=0,
                   release_country=None,release_date=None,stars=None,writers=None):
    # this function is used to input a attribut of a movie, and output recommand movie list
    # the fancy part is that you can use more than one attribut to obtain a recommand movie list
    recommend_list = []
    for i in dict_.keys():
        # 
        if director:
            if director in dict_[i]['director']:
                pass
            else:
                continue
        else:
            pass
        if genres:
            if genres in dict_[i]['genres']:
                pass
            else:
                continue
        else:
            pass
        if keywords:
            if keywords in dict_[i]['keywords']:
                pass
            else:
                continue
        else:
            pass
        if rating_value:
            if dict_[i]['rating_value'] >= rating_value:
                pass
            else:
                continue
        else:
            pass
        if release_country:
            if release_country in dict_[i]['release_country']:
                pass
            else:
                continue
        else:
            pass
        if release_date:
            if dict_[i]['release_date'] == release_date:
                pass
            else:
                continue
        else:
            pass
        if writers:
            if writers in dict_[i]['writers']:
                pass
            else:
                continue
        else:
            pass
        if stars:
            if stars in dict_[i]['stars']:
                pass
            else:
                continue
        else:
            pass
        dict_[i]['movie_name'] = i
        recommend_list.append(dict_[i])
    if not recommend_list:
        print('No search found. Please try again!')
    else:
        sorted_recommend_list = sorted(recommend_list,key=lambda x: x['rating_value'],reverse=True)
        return sorted_recommend_list
```

```python
def genreal_analysis_part1(new_dict,genre_name):
    # the choice of genre_name can be any value of
    # ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family',
    # 'Fantasy', 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller',
    # 'War', 'Western']
    
    # and this function will return a distribution of rating values by genre
    try:
        b = not_movie_name(new_dict,genres=genre_name)
        c = pd.DataFrame(b)
        c.set_index('movie_name',inplace = True)
    except:
        print('No search found. Please try again!')
        
    data = c[c['rating_value']>0]['rating_value']
    plt.hist(data.values, bins=20, density=True, facecolor="green", edgecolor="black", alpha=0.7,range=(0,10))
    plt.xlabel('rating_value')
    plt.title(f'{genre_name}')
    plt.show()
```

```python
def general_analysis_part2(new_dict,genre_name):
    # the choice of genre_name can be any value of
    # ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family',
    # 'Fantasy', 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller',
    # 'War', 'Western']
    
    # and this function will return a pie-chart describing the distribution of the released countries
    try:
        b = not_movie_name(new_dict,genres=genre_name)
        c = pd.DataFrame(b)
        c.set_index('movie_name',inplace = True)
    except:
        print('No search found. Please try again!')
        
    data = c['release_country']
    new_list = []
    for i in data.values:
        new_list.extend(i)
    new_df = pd.DataFrame(new_list,columns=['country'])
    new_arr = np.array([1]*44021)
    new_df['count'] = new_arr
    new_df.groupby('country').count() # obtain how many movies are released in each country
    stat_data = new_df.groupby('country').count().drop('a') 
    stat_data = stat_data.drop('N')
    stat_data.sort_values('count',inplace=True) # data preparation for drawing pie-chart
    
    # draw pit-chart
    colors = cm.rainbow(np.linspace(0, 1, 100))
    fig, ax = plt.subplots(figsize=(10,10))
    plt.pie(stat_data['count'], labels=stat_data.index, labeldistance=1.1, shadow=False, pctdistance=0.6,colors=colors)
    plt.title(f'{genre_name}')
    plt.show()
```

```python
def general_analysis_part3(new_dict,genre_name,country):
    # the choice of genre_name can be any value of
    # ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family',
    # 'Fantasy', 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller',
    # 'War', 'Western']
    
    # and this function will return a distribution of rating values of specified release country by genre 
    try:
        b = not_movie_name(new_dict,genres=genre_name,release_country=country)
        c = pd.DataFrame(b)
        c.set_index('movie_name',inplace = True)
    except:
        print('No search found. Please try again!')
        
    data = c[c['rating_value']>0]['rating_value']
    plt.hist(data.values, bins=20, density=True, facecolor="green", edgecolor="black", alpha=0.7,range=(0,10))
    plt.xlabel('rating_value')
    plt.title(f'{genre_name} movies released in {country}')
    plt.show()
```
